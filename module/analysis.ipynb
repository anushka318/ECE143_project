```python
import csv
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import os
import categorize_data as cd
import visualization as vz
import seaborn as sns
```

```python
current_directory = os.getcwd()
print(current_directory)
data = pd.read_csv('./clean_data.csv')
data = cd.categorize_data(data)    
```

```python
vz.plot_barplot(data=data, x='Anxiety_level', y='SWL_T', title='Anxiety vs Satisfaction', xlabel='Anxiety Level', ylabel='Satisfaction Level', save_as='anxiety_satisfaction.png')  
```

```python
# Heatmap: Anxiety levels based on Age Group & Employment
pivot_table = data.pivot_table(values='GAD_T', index='Age_group', columns='Work', aggfunc='mean')
plt.figure(figsize=(10, 5))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mean Anxiety Scores by Age Group and Employment")
plt.xlabel("Employment Status")
plt.ylabel("Age Group")
plt.show()
```

```python

# Aggregate mean Satisfaction with Life (SWL_T) by Degree category
swl_by_degree = data.groupby("Degree")["SWL_T"].mean().reset_index()

# Sort by satisfaction scores for better visualization
swl_by_degree = swl_by_degree.sort_values(by="SWL_T", ascending=False)
# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x="SWL_T", y="Degree", data=swl_by_degree, palette="coolwarm")
# Add title and labels
plt.xlabel("Average Satisfaction with Life (SWL_T)")
plt.ylabel("Degree")
plt.title("Mean Satisfaction with Life by Degree Level")

# Show the plot
plt.show()
```
