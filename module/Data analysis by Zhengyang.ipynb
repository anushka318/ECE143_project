```python
df = pd.read_csv("C:/Program Files/R Files/anxiety.csv", encoding = 'ISO-8859-1')
import numpy as np
import pandas as pd
import plotly.graph_objects as go
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
corr_masked = pd.DataFrame(corr).where(~mask, np.nan)  
size = max(600, min(50 * len(corr.columns), 1000))
fig = go.Figure(data=go.Heatmap(
    z=corr_masked,
    x=corr.columns,
    y=corr.columns,
    colorscale='Viridis',
    zmin=-1,
    zmax=1,
    zmid=0,
    colorbar=dict(
        len=0.8, 
        thickness=18, 
        title="Correlation",
        ticks="outside",
        ticklen=5
    )
))
fig.update_layout(
    title=dict(
        text='Correlation Matrix Heatmap',
        font=dict(size=18),
        x=0.5
    ),
    xaxis=dict(
        tickangle=-45,
        showgrid=False,
        zeroline=False
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False
    ),
    plot_bgcolor="white",
    width=size,
    height=size,
    template='plotly_white'
)
fig.show()
```

```python
print(df.Age.describe())
```

```python
import plotly.express as px
age_counts = df['Age'].value_counts().reset_index()
age_counts.columns = ['Age', 'Count']
top_n = 10
if len(age_counts) > top_n:
    other_count = age_counts.iloc[top_n:]['Count'].sum()
    age_counts = age_counts.iloc[:top_n]
    age_counts.loc[len(age_counts)] = ['Other', other_count] 
fig = px.pie(
    age_counts, names='Age', values='Count', 
    title='Age Distribution', 
    hole=0.3,  
    color_discrete_sequence=px.colors.qualitative.Set2 
)
fig.update_traces(
    textinfo='percent+label',
    textposition='inside',
    insidetextorientation='radial'  
)
fig.update_layout(
    width=900, 
    height=700, 
    margin=dict(l=50, r=50, t=50, b=50)
)
fig.show()
```

```python
import plotly.express as px
gender_counts = df['Gender'].value_counts()
total_count = gender_counts.sum()
percentages = (gender_counts / total_count * 100).round(2)
labels = [f"{count} ({percent}%)" for count, percent in zip(gender_counts.values, percentages)]
fig = px.bar(
    x=gender_counts.index, 
    y=gender_counts.values, 
    title='Count of Genders',
    labels={'x': 'Gender', 'y': 'Count'},
    text=labels 
)
fig.update_traces(marker_color=['#636EFA', '#EF553B'], textposition='outside')
fig.update_layout(
    width=700,
    height=500,
    xaxis_title='Gender',
    yaxis_title='Count',
    template='plotly_white'
)
fig.show()
```

```python
country_counts = df['Birthplace_ISO3'].value_counts().rename_axis('country').reset_index(name='count')
country_counts
```

```python
import plotly.express as px
df['Birthplace_ISO3'] = df['Birthplace_ISO3'].replace({'US': 'USA', 'U.S.': 'USA'})
top_countries = country_counts.sort_values(by='count', ascending=False).head(50)
fig = px.scatter_geo(
    data_frame=top_countries,
    locations='country',
    locationmode="ISO-3",
    size='count',
    title="Top Countries by Count",
    projection="natural earth",
    color='count',
    color_continuous_scale='Plasma',
    hover_name='country',
    hover_data={'count': True},
    size_max=50
)
fig.update_traces(marker=dict(line=dict(width=2, color='black')))
fig.show()
```

```python
games_counts = df['Game'].value_counts().rename_axis('Game').reset_index(name='count')
games_counts
```

```python
import plotly.express as px
games_counts['Percentage'] = (games_counts['count'] / games_counts['count'].sum() * 100).round(2)
top_n = 20 
games_counts_top = games_counts.head(top_n)
games_counts_top['Label'] = games_counts_top.apply(lambda row: f"{row['count']} ({row['Percentage']}%)", axis=1)
max_count = games_counts_top['count'].max()
games_counts_top['TextPosition'] = games_counts_top['count'].apply(lambda x: 'inside' if x > 0.5 * max_count else 'outside')
fig = px.bar(
    games_counts_top,
    x='count', 
    y='Game',  
    orientation='h',
    title='Top Games by Count',
    labels={'count': 'Count', 'Game': 'Game'},
    text='Label' 
)
fig.update_traces(
    marker_color='royalblue', 
    textposition=games_counts_top['TextPosition'],
    insidetextanchor='start',
    textfont_size=12 
)
fig.update_layout(
    width=1200, 
    height=900, 
    xaxis_title='Count',
    yaxis_title='Game',
    template='plotly_white',
    margin=dict(l=350, r=100, t=50, b=50),
    yaxis=dict(categoryorder="total ascending", tickfont=dict(size=12))
)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
max_hours_week = 16 * 7
legit_hour_df = df.query('Hours <= @max_hours_week')
residence_hours = legit_hour_df.groupby('Residence').Hours.agg(['mean', 'size']).query('Residence != "Unknown" & size >= 100')
residence_hours.sort_values(by=['mean'], ascending=[False], inplace=True)
plt.figure(figsize=(15, 6))
sns.boxplot(data=legit_hour_df[legit_hour_df['Residence'].isin(residence_hours.index)], 
            x='Hours', y='Residence', showfliers=False, palette="coolwarm")
plt.title("Distribution of Hours Played by Residence")
plt.xlabel("Hours Played")
plt.ylabel("Residence")
plt.show()
```

```python
import plotly.figure_factory as ff
fig = ff.create_2d_density(
    x=legit_hour_df["Age"], 
    y=legit_hour_df["Hours"], 
    colorscale="Blues", 
    hist_color="rgba(0,0,255,0.5)"
)
fig.update_layout(
    title="Age vs. Hours Played (Density Heatmap)",
    xaxis_title="Age",
    yaxis_title="Hours Played",
    width=800,
    height=600,
    template="plotly_white"
)
fig.show()
```

```python
import pandas as pd
import plotly.express as px
playstyle_choices = [
    'Multiplayer - online - with real life friends',
    'Multiplayer - online - with strangers', 
    'Multiplayer - online - with online acquaintances or teammates',
    'Singleplayer',
    'Multiplayer - offline (people in the same room)'
]
playstyle_counts = df['Playstyle'].value_counts().reindex(playstyle_choices, fill_value=0)
total_count = playstyle_counts.sum()
percentages = (playstyle_counts / total_count * 100).round(2) 
playstyle_counts_df = pd.DataFrame({
    'Playstyle': playstyle_counts.index,
    'Count': playstyle_counts.values,
    'Percentage': percentages.values
})
playstyle_counts_df['Label'] = playstyle_counts_df.apply(lambda row: f"{row['Count']} ({row['Percentage']}%)", axis=1)
fig = px.bar(playstyle_counts_df, 
             x='Count', y='Playstyle', 
             orientation='h', 
             title='Distribution of Playstyle Choices',
             text='Label', 
             color='Playstyle', 
             color_discrete_sequence=px.colors.qualitative.Set2)
fig.update_layout(
    xaxis_title="Count",
    yaxis_title="Playstyle",
    width=1200,
    height=600,  
    template="plotly_white",
    margin=dict(l=150, r=50, t=50, b=50), 
    yaxis=dict(tickmode="array", tickfont=dict(size=14)), 
)
fig.update_traces(textposition='outside')
fig.show()
```
